```python
import csv
import random
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns # for data visualization
import pandas as pd # for data analysis
import numpy as np

filename = "deliveries.csv"
fields = []
rows = []

with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    
    fields = next(csvreader)
    
    for row in csvreader:
        rows.append(row)
    print("Total no. of rows: %d"%(csvreader.line_num)) 

print('Field names are:' + ', '.join(field for field in fields))

rowso=list()
fields0=list()
fields0.append('MATCH')
for x in range(1,21):
    fields0.append(x)
    
minn=0
match=''
fig = plt.figure(figsize= (16,16))
ax1 = fig.add_subplot(111)

# x-axis label 

for e in range(0,8):
    flag=0
    over=list()
    runs=list()
    min1=minn
   # print(min1)
    flag2=0
    for i in range(min1,2000) :
        
            
        
        if((int(rows[i][4])==20) ):
            #print('rows[i][4]!=1',int(rows[i][4])!=1)
           # print('rows[i+1][4]==1',int(rows[i+1][4])==1)
           # print('(rows[i][4]!=1 and rows[i+1][4]==1)',(int(rows[i][4])!=1 and int(rows[i+1][4])==1))
            #print('rows[i][4]',rows[i][4],'rows[i+1][4]',rows[i+1][4])
            match=rows[i][0]+rows[i][1]+rows[i][2]
            flag=1
        if((int(rows[i][4])!=1 and int(rows[i+1][4])==1)):
            match='Match'+rows[i][0]+' INNINGS '+rows[i][1]+' TEAM '+rows[i][2]
            flag2=1
            
        if(flag==1 and int(rows[i][4])!=20):
             break
        if(flag2==1 and int(rows[i][4])==1):
            break
            
        #print("%5s"%row[4],end='')
        over.append(int(rows[i][4]))
        
        #print("%5s"%row[17])
        runs.append(int(rows[i][17]))
        minn+=1
        #print('I',i);
    #print('overs',over)
   # print('runs',runs)
    
    x=list()
    y=list()
    i=1
    total=int(runs[0])
    x.append(int(over[0]))
    while(i<len(over)):
        if(over[i]!=over[i-1]):
            x.append(int(over[i]))
            y.append(int(total))
            total=int(runs[i])
        else:
            total=total+int(runs[i])
            if(i==len(over)-1):
                y.append(total)
        i+=1
    
    print("runs",y)
    print("overs",x)
    # plotting points as a scatter plot 
    r = random.random()
    b = random.random()
    g = random.random()
    color = (r, g, b)
   
    ax1.scatter(x, y, s=140,label=match)
   
# function to show the plot 
plt.yticks(np.arange(0, 40, 1))
plt.xticks(np.arange(0, 21, 1))
plt.xlabel('overs',fontsize=20) 
    # frequency label 
plt.ylabel('runs',fontsize=20) 
    # plot title 
plt.title('runs scored by the teams',fontsize=25) 
    # showing legend 
plt.legend() 
plt.show() 
```
